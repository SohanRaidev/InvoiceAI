```python
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
from sklearn.ensemble import IsolationForest
from transformers import pipeline
import pandas as pd

# Initialize FastAPI app
app = FastAPI()

# Load multilingual BERT model for text processing
nlp = pipeline("feature-extraction", model="bert-base-multilingual-cased")

# Initialize Isolation Forest for fraud detection
fraud_detector = IsolationForest(contamination=0.01)  # Adjust contamination as needed

# Define request model
class Invoice(BaseModel):
    text: str  # Invoice text
    amount: float  # Invoice amount

class BatchInvoices(BaseModel):
    invoices: list[Invoice]  # List of invoices

# Endpoint for single invoice processing
@app.post("/process-invoice")
def process_invoice(invoice: Invoice):
    try:
        # Extract features using mBERT
        features = nlp(invoice.text)

        # Detect fraud using Isolation Forest
        is_fraud = fraud_detector.predict([[invoice.amount]])[0] == -1

        return {
            "text_features": features,
            "is_fraud": bool(is_fraud)
        }
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

# Endpoint for batch invoice processing
@app.post("/process-batch")
def process_batch(batch: BatchInvoices):
    try:
        results = []
        for invoice in batch.invoices:
            features = nlp(invoice.text)
            is_fraud = fraud_detector.predict([[invoice.amount]])[0] == -1
            results.append({
                "text": invoice.text,
                "amount": invoice.amount,
                "is_fraud": bool(is_fraud)
            })
        return {"results": results}
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

# Run the FastAPI app
if __name__ == "__main__":
    import uvicorn
    uvicorn.run(app, host="0.0.0.0", port=8000)
```

```python
import streamlit as st
import requests

# Streamlit app title
st.title("Invoice Processing System")

# Input fields for single invoice
st.header("Process Single Invoice")
invoice_text = st.text_area("Enter Invoice Text")
invoice_amount = st.number_input("Enter Invoice Amount", value=0.0)

# Process single invoice
if st.button("Process Single Invoice"):
    response = requests.post(
        "http://localhost:8000/process-invoice",
        json={"text": invoice_text, "amount": invoice_amount}
    )
    if response.status_code == 200:
        result = response.json()
        st.write("Text Features:", result["text_features"])
        st.write("Is Fraud:", result["is_fraud"])
    else:
        st.error("Error processing invoice")

# Input fields for batch processing
st.header("Process Batch Invoices")
batch_invoices = st.text_area("Enter Batch Invoices (JSON format)", value='[{"text": "Sample Invoice 1", "amount": 100.0}, {"text": "Sample Invoice 2", "amount": 1000.0}]')

# Process batch invoices
if st.button("Process Batch Invoices"):
    try:
        batch = eval(batch_invoices)  # Convert JSON string to list of dictionaries
        response = requests.post(
            "http://localhost:8000/process-batch",
            json={"invoices": batch}
        )
        if response.status_code == 200:
            results = response.json()["results"]
            st.write("Batch Results:", results)
        else:
            st.error("Error processing batch")
    except Exception as e:
        st.error(f"Invalid input format: {e}")
```
